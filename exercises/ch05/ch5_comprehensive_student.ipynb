## 発展問題　Precision / Recall を計算する ★★★

> ⏰ **時間が余った人だけ取り組んでください（AI 使用可）**

### ❓ なぜ正解率だけでは足りないのか

> **「正解率が高くても、悪性の見逃しが多い」ことがあります。**

| 指標 | 意味 | 医療での重要性 |
|------|------|--------------|
| Precision（適合率） | 「悪性」と予測した中で、本当に悪性の割合 | 誤検知を減らしたいとき |
| Recall（再現率） | 実際の悪性の中で、正しく「悪性」と予測できた割合 | **見落としを減らしたいとき** |

> 医療診断では **Recall を高く** したい（悪性を見逃してはならない）。

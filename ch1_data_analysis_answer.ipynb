## 準備  ライブラリとデータを読み込む

⛔ 注意 最初にこのセルを実行してください。実行しないと以降が動きません。

````python
import pandas as pd
from sklearn.datasets import load_wine

wine = load_wine()
df   = pd.DataFrame(wine.data, columns=wine.feature_names)
df["品種名"] = ["wine_" + str(t) for t in wine.target]
print(f"行数: {len(df)}  列数: {len(df.columns)}")
````

## 問2  品種ごとにアルコールとプロリンの平均を比べる ★★☆

実務でのイメージ：
「部署別の売上平均を比べる」「商品カテゴリ別の単価の違いを調べる」
-- グループ別の集計は、差の発見と原因特定に使う基本技術です。

---

Q2-1：品種ごとの「アルコール度数の平均」を調べてみましょう

3品種のアルコール度数（alcohol）の平均値を、品種別に比較してください。

💡 ヒント：「グループごとに特定の列の平均を計算する方法」を Copilot に聞いてみましょう。

````python
# 品種ごとに alcohol の平均を集計して比較する
df.groupby("品種名")["alcohol"].mean().round(2)
````

---
Q2-2：今度は alcohol と proline の両方を同時に集計してみましょう

品種ごとに 2 列まとめて平均を出してください。

📌 なぜ proline？：Q1-3 の `describe()` で std（標準偏差）が最も大きかった列です。  
「ばらつきが大きい = 品種ごとに差が出やすい」可能性が高いため、alcohol とセットで確認します。

💡 ヒント：「複数の列をグループ別に同時集計する方法」を Copilot に聞いてみましょう。

````python
# 品種ごとに alcohol と proline の平均を一度にまとめて集計する
df.groupby("品種名")[["alcohol", "proline"]].mean().round(2)
````

---
Q2-3（追加）  平均だけでなく「最大・最小」も見てみましょう

平均（mean）だけでは「ばらつきの大きさ」「極端な値の有無」がわかりません。
最大・最小・中央値を同時に確認すると、品種の特徴がより鮮明になります。

📌 例えば wine_0 の proline の最小値は？wine_1 の最大値より低いですか？

💡 ヒント：「groupby と agg を組み合わせて複数の統計量を同時に計算する方法」を Copilot に聞いてみましょう。

````python
# 品種ごとに alcohol と proline の平均・最大・最小を一度に計算して、品種間の差をより詳しく確認する
df.groupby("品種名")[["alcohol", "proline"]].agg(["mean", "max", "min"]).round(1)
````

## 問4（発展）  条件でデータを絞り込む ★★★

📋 発展 時間が余った人だけ取り組んでください

実務でのイメージ：
「アルコール度数が高いワインだけを抽出して品種の内訳を確認する」
← 「売上が前月比 -20% 以下の商品だけ抽出して原因を調べる」と同じ操作です。

---

Q：アルコール度数が高く、かつプロリンも多いワインだけを取り出してみましょう

「アルコール >= 13.5 かつ プロリン >= 700」の条件で絞り込み、以下を確認してください。

1. 絞り込んだデータは何件ありましたか？
2. 品種（wine_0・wine_1・wine_2）の内訳はどうなっていますか？

💡 なぜこの数値？：問2の集計結果で wine_0 の alcohol 平均が 13.74、proline 平均が 1115 と他の品種より大幅に高かったことを参考に、「wine_0 っぽい条件」を設定しています。

💡 ヒント：「2つの条件を同時に満たす行だけを絞り込む方法」を Copilot に聞いてみましょう。

````python
# AND 条件（アルコール高 かつ プロリン多）でデータを絞り込み、品種の内訳を確認する
premium = df[(df["alcohol"] >= 13.5) & (df["proline"] >= 700)]
print(f"絞り込み後: {len(premium)} 件")
premium["品種名"].value_counts()
````

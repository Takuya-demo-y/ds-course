## ✅ 1日のまとめ：データサイエンティストの仕事の流れ

```
1. データを読み込む（pandas）
      ↓
2. データの中身を把握する（EDA：可視化・集計）
      ↓
3. 仮説を立てる（「この変数が重要なはず」）
      ↓
4. モデルを作って検証する（scikit-learn）
      ↓
5. 結果を解釈して、ビジネスに活かす
```

### 生成AIとの協業まとめ

| 役割 | 誰が担う |
|------|----------|
| コードの書き方を調べる | 生成AI |
| エラーを修正する | 生成AI |
| **何を分析するかを決める** | **自分** |
| **数値・グラフを解釈する** | **自分** |
| **ビジネス上の示唆を出す** | **自分** |

生成AIはあくまで「コーディングの補助ツール」。
**分析の目的設定と結果の解釈は、データサイエンティスト（あなた）の仕事です。**

### 今日覚えてほしい3つのこと

1. **データサイエンティストの仕事の大半はEDA（データ理解）**  
   モデルを作ることより、データを理解することの方が重要で時間もかかります。

2. **生成AIはコードを書く「道具」、分析の「目的と解釈」はあなたの仕事**  
   AIに「分析して」と丸投げするのではなく、「何を知りたいか」を自分で決めてから使う。

3. **「完全に理解してから使う」より「使いながら理解する」**  
   コードが全部わからなくても、実行して結果を見て、解釈して進める。それで十分です。

## 2. データサイエンティストの仕事とは

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import japanize_matplotlib

fig, ax = plt.subplots(figsize=(13, 3.5))
ax.axis('off')

steps = [
    ('① データ収集\n・ 準備', '#4A90D9'),
    ('② EDA\n（探索的分析）', '#5BA85A'),
    ('③ 可視化\n・ 仮説立て', '#E8A838'),
    ('④ モデル構築\n・ 評価', '#C0392B'),
    ('⑤ 結果の解釈\n・ 報告', '#8E44AD'),
]

x_positions = [0.08, 0.26, 0.44, 0.62, 0.80]
for (label, color), x in zip(steps, x_positions):
    fancy = mpatches.FancyBboxPatch((x - 0.085, 0.15), 0.155, 0.7,
                                    boxstyle='round,pad=0.02',
                                    facecolor=color, edgecolor='white',
                                    linewidth=2, transform=ax.transAxes)
    ax.add_patch(fancy)
    ax.text(x, 0.50, label, transform=ax.transAxes,
            ha='center', va='center', fontsize=11, color='white', fontweight='bold')

# 矢印
for x in [0.18, 0.36, 0.54, 0.72]:
    ax.annotate('', xy=(x + 0.005, 0.50), xytext=(x - 0.005, 0.50),
                xycoords='axes fraction', textcoords='axes fraction',
                arrowprops=dict(arrowstyle='->', color='#333333', lw=2))

# 今日扱うハイライト
ax.text(0.44, 0.04, '← 今日体験する範囲 →', transform=ax.transAxes,
        ha='center', va='bottom', fontsize=10, color='#555555',
        style='italic')
ax.annotate('', xy=(0.715, 0.08), xytext=(0.165, 0.08),
            xycoords='axes fraction', textcoords='axes fraction',
            arrowprops=dict(arrowstyle='<->', color='#555555', lw=1.5))

ax.set_title('データサイエンティストの仕事のフロー', fontsize=14, pad=15)
plt.tight_layout()
plt.show()
````

## 3. 今日の流れ

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(13, 4))
ax.axis('off')

schedule = [
    (0,   1.5, 'オリエン\nテーション', '#95A5A6', '9:00\n〜9:20'),
    (1.5, 3,   'Ch.1\nデータ\n読み込み', '#4A90D9', '9:20\n〜10:20'),
    (4.5, 3,   'Ch.2\n可視化\nEDA', '#5BA85A', '10:30\n〜11:30'),
    (7.5, 3,   'Ch.3\n画像処理', '#E8A838', '11:30\n〜12:30'),
    (11.5,3,   'Ch.4\n機械学習', '#C0392B', '13:30\n〜14:30'),
    (15,  4.5, 'Ch.5\n総合演習', '#8E44AD', '14:40\n〜16:10'),
    (19.5,1.5, 'まとめ\n振り返り', '#95A5A6', '16:10\n〜16:30'),
]

breaks = [(3, 1.5, '休憩'), (10.5, 1, '昼食'), (14.5, 0.5, '休憩')]

total_w = 21
bar_y, bar_h = 0.5, 0.35

for start, width, label, color, time in schedule:
    x = start / total_w
    w = width / total_w
    rect = mpatches.FancyBboxPatch((x + 0.002, bar_y), w - 0.004, bar_h,
                                    boxstyle='round,pad=0.005',
                                    facecolor=color, edgecolor='white', lw=1.5,
                                    transform=ax.transAxes)
    ax.add_patch(rect)
    ax.text(x + w/2, bar_y + bar_h/2, label, transform=ax.transAxes,
            ha='center', va='center', fontsize=9, color='white', fontweight='bold')
    ax.text(x + w/2, bar_y - 0.08, time, transform=ax.transAxes,
            ha='center', va='top', fontsize=7.5, color='#555555')

for start, width, label in breaks:
    x = start / total_w
    w = width / total_w
    rect = mpatches.FancyBboxPatch((x + 0.002, bar_y), w - 0.004, bar_h,
                                    boxstyle='round,pad=0.005',
                                    facecolor='#EAECEE', edgecolor='#BDC3C7', lw=1,
                                    transform=ax.transAxes)
    ax.add_patch(rect)
    ax.text(x + w/2, bar_y + bar_h/2, label, transform=ax.transAxes,
            ha='center', va='center', fontsize=8, color='#888888')

ax.set_title('1日のスケジュール', fontsize=14, pad=25)
plt.tight_layout()
plt.show()
````

## 4. 生成AIの使い方ルール（本日の共通ルール）

> **講師メモ**：ここは全員で合意する最重要ポイント。「AIのコードが多すぎてわからない」を防ぐための設計です。

### 生成AIは「賢いコーディング補助ツール」

```
自分の仕事 ＝ 「何を分析したいか」を決める ＋ 「結果が何を意味するか」を解釈する
AIの仕事  ＝ 「どうコードを書くか」を教える
```

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(11, 4.5))
ax.axis('off')

left_items = [
    'コードの書き方がわからない',
    'ライブラリのオプションを調べたい',
    'エラーメッセージを修正したい',
    'グラフの見た目を整えたい',
]
right_items = [
    '何を分析するかを決める',
    'グラフから仮説を立てる',
    '欠損値の処理方針を判断する',
    'モデルの結果をビジネスに結びつける',
]

# 左ボックス（AIに任せる）
left_box = mpatches.FancyBboxPatch((0.02, 0.05), 0.42, 0.88,
    boxstyle='round,pad=0.02', facecolor='#EBF5FB', edgecolor='#4A90D9', lw=2,
    transform=ax.transAxes)
ax.add_patch(left_box)
ax.text(0.23, 0.90, '✅ AIに任せてよいこと', transform=ax.transAxes,
        ha='center', fontsize=12, fontweight='bold', color='#2471A3')
for i, item in enumerate(left_items):
    ax.text(0.06, 0.74 - i*0.17, f'・ {item}', transform=ax.transAxes,
            va='top', fontsize=10, color='#1A5276')

# 右ボックス（自分でやる）
right_box = mpatches.FancyBboxPatch((0.56, 0.05), 0.42, 0.88,
    boxstyle='round,pad=0.02', facecolor='#FEF9E7', edgecolor='#E8A838', lw=2,
    transform=ax.transAxes)
ax.add_patch(right_box)
ax.text(0.77, 0.90, '❌ 自分でやること', transform=ax.transAxes,
        ha='center', fontsize=12, fontweight='bold', color='#9A7D0A')
for i, item in enumerate(right_items):
    ax.text(0.59, 0.74 - i*0.17, f'・ {item}', transform=ax.transAxes,
            va='top', fontsize=10, color='#7D6608')

# 中央矢印
ax.text(0.50, 0.50, '⇆', transform=ax.transAxes,
        ha='center', va='center', fontsize=22, color='#555555')

ax.set_title('生成AIの活用ルール', fontsize=14, pad=12)
plt.tight_layout()
plt.show()
````
